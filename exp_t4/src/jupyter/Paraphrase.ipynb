```python
file = "../data/COLIEE2024statute_data-English/text/civil_code_en-1to724-2.txt"
# f = open(file, "r")
def sub_articles(text):
    text = text.split("\n")
    c_result = []
    result = ""
    for para in text:
        if ". (" in para:
            tmp = para.split(". (")
            for data in tmp:
                if data[0].isdigit():
                    data = "(" + data
                if data[-1] != ".":
                    data = data + "."
                c_result.append(data)
        else:
            c_result.append(para)

    if len(c_result) > 1:
        result = ""
        for i in range(len(c_result)-1):
            result += c_result[i]
            if c_result[i+1][1].isdigit() == False:
                result += " "
            else:
                 result += "\n"
        result += c_result[-1]
        c_result = result.strip().split("\n")
    return c_result
```

```python
def preprocess(data):
    final_check = []
    for item in data:
        arts = item["result"].split("\n")
        count_d = 0
        key = ""
        test_articles = {}
        text_aux = ""
        # print(arts)
        for para in arts:
            result = []
            if para.split()[0] == "Article":
                count_d = 0
                para = para.split()
                key = " ".join(para[:2]).strip()
                test_articles.update({key: {}})
                text = " ".join(para[2:]).strip()
                result = sub_articles(text)
                for text in result:
                    count_d += 1
                    test_articles[key].update({count_d: text})
            else:
                if para[1].isdigit() == False:
                    continue
                else:
                    result = sub_articles(para)
                    if key != "":
                        for text in result:
                            count_d += 1
                            test_articles[key].update({count_d: text})
        item.update({"rewrite": test_articles})
        final_check.append(item)
    return final_check
```

```python
def extract_rule(text):
    result = []
    text = text.split("\n")
    for clause in text:
        if clause == "" or "none" in clause.lower(): continue
        else: 
            text = clause.replace("provided, however,", "However").split(": ")
            if "condition" in text[0].lower():
                # print(text)
                # print("*******************")
                text = text[-1].replace('"', "")
                if text[:2].strip().lower() == "if":
                    text = text.capitalize() 
                elif text.split()[0] != "However": 
                    text = "If "+text[0].lower() + text[1:]
            else:
                text = text[-1]
            result.append(text)
            
    return " ".join(result).strip()

# def get_article(text):
    
```

```python
for file in files:
    final_data = []
    data = load_jsonl(file)
    for item in data:
        rewrite = item["rewrite"]
        result = ""
        for key in rewrite:
            result += key+"\n"
            if key not in articles:
                print(key)
                print(item["index"])
                print(file)
            else:
                count = 0
                for k in rewrite[key]:
                    count += 1
                    if k in articles[key]:
                        result += str(count) + ". "+extract_rule(articles[key][k]) + "\n"
        result = result.strip()
        print(result)   
        # print(item["result"])
        print("====================")
        if "(i)" in item["result"]:
            result = item["result"]
        item.update({"rewrite": result})
        final_data.append(item)
    file = file.split("/")[-1]
    with jsonlines.open("../data/COLIEE2024statute_data-English/fewshot.json/rewrite/"+file, mode='w') as writer:
        writer.write_all(final_data)
```
